```python
from pinecone import Pinecone

# Replace with your actual Pinecone API key
pc = Pinecone(api_key='# Replace with you actual Pinecone API key')

# Create a Pinecone index if it doesn't exist
index_name = 'llama-index-nvidia'
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=1536,
        metric='cosine'
    )

# Connect to the Pinecone index
pinecone_index = pc.Index(index_name)
```

```python
# Initialize the Pinecone vector store
vector_store = PineconeVectorStore(pinecone_index=pinecone_index)

# Create a storage context using the vector store
storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
import nest_asyncio
nest_asyncio.apply()

from llama_index.readers.github import GithubClient, GithubRepositoryReader
import asyncio

# Initialize the GitHub client
github_client = GithubClient("# Replace with your actual token")  # Replace with your actual token

# Initialize the loader with your GitHub repository details
loader = GithubRepositoryReader(
    github_client=github_client,
    owner='NVIDIA',
    repo='GenerativeAIExamples',
    use_parser=False,
    verbose=True,
    filter_directories=(
        ['.git', '__pycache__'], GithubRepositoryReader.FilterType.EXCLUDE
    ),
    filter_file_extensions=(
        ['.md', '.txt', '.py', '.java', '.js', '.html', '.css'], GithubRepositoryReader.FilterType.INCLUDE
    ),
)

# Define an async function to load the data
async def load_data():
    return await loader.aload_data(branch='main')

# Run the async function
documents = asyncio.get_event_loop().run_until_complete(load_data())
```

```python
# Create the index using the documents and the storage context
index = VectorStoreIndex.from_documents(documents, storage_context=storage_context)
```

```python
# Create a query engine
query_engine = index.as_query_engine()

# Perform a query
response = query_engine.query('Please tell me about Nvidia NEMO')
print(response)
```
